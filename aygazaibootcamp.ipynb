```python

#  kütüphaneleri içe aktarıyorum
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score

#  veri seti seçiyorum ve fashion mnist veri setini yüklüyorum
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# veri setini bölüyorum
print("X_train boyutu:", X_train.shape)
print("y_train boyutu:", y_train.shape)
print("X_test boyutu:", X_test.shape)
print("y_test boyutu:", y_test.shape)

#  görüntülerin boyutlarını yazdırıyorum
print("Görüntü boyutu:", X_train[0].shape)

#  görselleştirme
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title("Label: " + str(y_train[i]))
    plt.axis('off')
plt.show()

# verileri normalize ediyorum
X_train = X_train / 255.0
X_test = X_test / 255.0

# reshape işlemi: KNN modeli için verileri düzleştiriyoruz
X_train_flattened = X_train.reshape(len(X_train), -1)
X_test_flattened = X_test.reshape(len(X_test), -1)

# KNN Modeli oluşturma ve eğitme
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_flattened, y_train)

# modeli test verisi ile tahmin yapma
y_pred = knn.predict(X_test_flattened)

# Accuracy hesaplama
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# F1-Score hesaplama
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1-Score:", f1)

```
